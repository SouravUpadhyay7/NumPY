```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basic Line Plot

x = np.linspace(0,10,100)
y = np.cos(x)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
# Custom Line Graph

x = np.linspace(0,10,50)
y = np.cos(x)

plt.plot(x,y, 'o', linestyle = '--', color = 'red', markersize = 5, alpha = .5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
# Line plot with multiple lines

x = np.linspace(0,10,50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, 'o', linestyle = '--', color = 'red', markersize = 5, alpha = .5 , label = 'sin(x)')
plt.plot(x,y2, 'o', linestyle = '--', color = 'blue', markersize = 5, alpha = .5, label = 'cos(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
```

```python
#  Basic Scatter Plot
x = np.random.rand(500)
y = np.random.rand(500)

plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
plt.show()
```

```python
# Customized Scatter Plot


x = np.random.rand(500)
y = np.random.rand(500)
z = np.random.rand(500)

plt.scatter(x,y, marker = 'o', c = z, alpha = .7, cmap = 'viridis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
plt.colorbar()
plt.show()
```

```python
# Scatter Plot with Multiple Classes


x1 = np.random.rand(200) - .5
y1 = np.random.rand(200) - .5

x2 = np.random.rand(200) + .5
y2 = np.random.rand(200) + .5


plt.scatter(x1,y1, marker = 'o', alpha = .7)
plt.scatter(x2,y2, marker = 'o', alpha = .7)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.colorbar()
plt.show()
```

```python
# Hexbining Scatter Plot

x1 = np.random.rand(10000)
y1 = np.random.rand(10000)


plt.hexbin(x1,y1, gridsize = 20)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hexbining Plot')
plt.colorbar()
plt.show()
```

```python
# Basic Bar Graph


x = np.array(['A','B','C','D'])
y = np.array([10,20,30,40])

plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Bar Graph')
plt.show()
```

```python
# Stacked Bar Graph


data = np.random.rand(10,5)

plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3] + data[:,4], label = 'D-5')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3], label = 'D-4')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2], label = 'D-3')
plt.bar(np.arange(10), data[:,0] + data[:,1],label = 'D-2')
plt.bar(np.arange(10), data[:,0], label = 'D-1')

plt.legend()
plt.show()
```

```python
# Grouped Bar Graph


data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

width = .25
x = np.arange(3)

plt.bar(x - width, data[:,0], width,label = 'D-1')
plt.bar(x        , data[:,1], width, label = 'D-2')
plt.bar(x + width, data[:,2], width, label = 'D-3')

plt.legend()
plt.show()
```

```python
# Basic Histogram


data = np.random.randn(10000000)

plt.hist(data, bins = 1000)
plt.show()
```

```python
# Cumulative Histogram

data = np.random.randn(1000000)

plt.hist(data, bins = 1000, cumulative = True)
plt.show()
```

```python
#  Histogram with Multiple Datasets

data1 = np.random.randn(100000)
data2 = np.random.randn(100000) + 2

plt.hist(data1, bins = 100, alpha = .7, label = 'D1')
plt.hist(data2, bins = 100, alpha = .7, label = 'D2')
plt.legend()
plt.show()
```

```python
# Box Plot


data = np.random.randn(100,5)

plt.boxplot(data, labels = ['D-1','D-2','D-3','D-4','D-5'], vert = False)
plt.show()
```

```python
# Heatmap

data = np.random.rand(10,10)

plt.imshow(data)
plt.colorbar()
plt.show()
```

```python
# Pie Chart


labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 9.2,     8.5,       7.7]
explode = (0.13,        .1,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']


plt.title("Programming Language Popularity")
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.0f%%', shadow=True, explode=explode ,startangle=140)

plt.axis('equal')
plt.legend(title="Languages", loc="upper left")

plt.show()
```
